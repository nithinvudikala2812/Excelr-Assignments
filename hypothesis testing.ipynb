#HYPOTHESIS TESTING

Data Provided:

•	The theoretical weekly operating cost model: W = $1,000 + $5X

•	Sample of 25 restaurants with a mean weekly cost of Rs. 3,050

•	Number of units produced in a week (X) follows a normal distribution with a mean (μ) of 600 units and a standard deviation (σ) of 25 units


Hypothesis statement

H0:mu=4000

H1:mu>4000

where 4000 is the theoretical mean weekly operating cost calculated from the model  W=$1,000+$5X  with  X=600 .

Test Statistics calculation:

given,

•	ˉxˉ = sample mean weekly cost (Rs. 3,050)

•	μ = theoretical mean weekly cost according to the cost model (W = $1,000 + $5X for X = 600 units)=4000

•	σ = 5*25 units=125 units

•	n = sample size (25 restaurants)

z_cal=(ˉxˉ-μ)/ (σ/sqrt n)
    
      



```python
from scipy.stats import norm
import math

# Given values
sample_mean = 3050
population_mean = 4000
population_std = 125
sample_size = 25

# Calculate z
z_cal = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))

# Right-tailed p-value
p_value = 1 - norm.cdf(z_cal)

print("Z calculated:", z_cal)
print("p-value (right-tailed):", p_value)
```

```python
#determining the critical value
import math
from scipy.stats import norm

alpha = 0.05 # Significance level

# Since this is a right-tailed test, we need the z-value for 1 - alpha
z_critical = norm.ppf(1 - alpha)

print("Critical z-value:", z_critical)
```

Determining critical value:

Therefore it falls under the accepted region.


```python
#making a decision
if z_cal > z_critical:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

Making a decision:

the null hypothesis is accepted based on the test statistics calculations.


Conclusion:

Based on the decision we can conclude that there is no stong evidence to prove restaurant owners' claim that the weekly operating costs are higher than the model suggests.
