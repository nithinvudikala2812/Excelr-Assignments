#Confidence Interval

```python
import numpy as np
data=np.array([[1.13], [1.55], [1.43], [0.92], [1.25], [1.36], [1.32], [0.85], [1.07], [1.48], [1.20], [1.33], [1.18], [1.22], [1.29]])
data
```

```python
import pandas as pd
df=pd.DataFrame(data, columns=['print_heads'])
df
```

```python
df.mean()
```

```python
df.std()
```

99% Confidence Interval Using Sample Standard Deviation

```python
import scipy.stats as st
import numpy as np

# Sample mean and standard deviation are already calculated in previous cells
sample_mean = df['print_heads'].mean()
sample_std = df['print_heads'].std()
n = len(df)
confidence_level = 0.99

# Degrees of freedom
degrees_freedom = n - 1

# Critical t-value
# We need the t-value for a two-tailed test, so we use (1 - confidence_level) / 2 for the alpha level in the ppf function.
alpha = (1 - confidence_level) / 2
critical_t_value = st.t.ppf(1 - alpha, degrees_freedom)

# Margin of error
margin_of_error = critical_t_value * (sample_std / np.sqrt(n))

# Confidence interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Degrees of Freedom: {degrees_freedom}")
print(f"Critical t-value for {confidence_level*100}% confidence: {critical_t_value:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"99% Confidence Interval(using sample std dev): ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f})")
```

we are using t distribution for this task because the total population data is only 15 which is N<30.

99% Confidence Interval Using Known Population Standard Deviation

```python
import scipy.stats as st
import numpy as np

# Sample mean is the same
sample_mean = df['print_heads'].mean()
n = len(df)
confidence_level = 0.99

# Population standard deviation (given as 0.2 million)
population_std = 0.2

alpha = (1 - confidence_level) / 2
critical_z_value = st.norm.ppf(1 - alpha)

# Margin of error
margin_of_error = critical_z_value * (population_std / np.sqrt(n))

# Confidence interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Population Standard Deviation: {population_std:.4f}")
print(f"Critical z-value for {confidence_level*100}% confidence: {critical_z_value:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"99% Confidence Interval (using population std dev): ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f})")
```
