#Descriptive Analytics for Numerical Columns

```python
#statistical measures for all the numerical columns
numerical_cols=["Volume","Avg Price","Total Sales Value","Discount Rate (%)","Discount Amount","Net Sales Value"]

#mean
df[numerical_cols].mean()
```

```python
#median
df[numerical_cols].median()
```

```python
#mode
df[numerical_cols].mode()
```

```python
#standard deviation
df[numerical_cols].std()
```

The above output displays the statistical measures for the numericals columns which are Volume,Avg Price,Total Sales Values,Discount Rate(%),Discount Amount,Net Sales Value.
Let us talk about mean,median,mode and standard deviation.

Mean:It is the average of the values in a numerical column which is calculated by dividing the sum of values by total number of values.

Median:It is the center value if the n value is odd and if n is even average of the middle two values will be the median.

Mode:The frequently repeated value will be the mode.

Standard Deviation:How far the data is being deviated from the center which is mean to the left and to the right.


#Data Visualization

Histograms

```python
import matplotlib.pyplot as plt

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    df[col].hist()
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

All the histograms of numerical variables has positive skewedness which means the outliers are on the left side except in the case of Discount Rate(%) which is negatively skewed that means the outliers will be on the left.We did not find any symmetrical shaped histogram hence the possibility of the outliers in the data is more.

Boxplots

```python
import matplotlib.pyplot as plt

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    df[col].plot(kind='box', vert=False)
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"Column: {col}")
    print(f"  IQR: {IQR}")
    print(f"  Lower Bound: {lower_bound}")
    print(f"  Upper Bound: {upper_bound}")
    print(f"  Outliers (using IQR method):")
    #percentage of outliers

    display(outliers[[col]]) # Display only the outlier values for clarity
    print(f"  Percentage of Outliers: {len(outliers) / len(df) * 100:.2f}%")
    print("-" * 30)
```

As we can see the percentage of outliers in each column is very high which means there is a unusual distribution in the data and it has been spread out very wide, the more the percentage of outliers the more it is spread and more errors in the data.

Bar chart analysis for categorical variables

```python
import matplotlib.pyplot as plt

categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

The above are the bar graphs for all the categorical variables.From my analysis i noticed that the frequency of data in every categorical variable is same or almost same.

Standardization of Numerical Variables

Standardization:This will removes the units from the numerical columns after performing the standardization function using the formula z=x-mu/sigma, it will now be within range of -3 to +3.

```python
df.info()
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
df[numerical_cols]=pd.DataFrame(df[numerical_cols])
df[numerical_cols]
```

Earlier the data is widely distibuted now after standardization or z_score normalization the data is now distributed within the range of -3 to +3 for all the numerical variables.

Conversion of Categorical Data into Dummy Variables

Categorical data is converted into dummy variables because all the regression models only work with numerical variables.One hot encoding converts the qualitative data into quantitative form that the models can understand.

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
categorical_cols = df.select_dtypes(include=['object']).columns
df_encoded = pd.DataFrame(ohe.fit_transform(df[categorical_cols]).toarray(), columns=ohe.get_feature_names_out(categorical_cols))
df_encoded.head()
```
