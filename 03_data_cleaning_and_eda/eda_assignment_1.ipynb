```python
#detect outliers
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers=df[(df<lower_bound)|(df>upper_bound)]
```

```python
# Recode outliers using the median
for column in df.columns:
  df[column] = np.where(
      (df[column] < lower_bound[column]) | (df[column] > upper_bound[column]),
      df[column].median(),
      df[column],
  )

# Display the updated DataFrame and info to confirm the changes
display(df)
df.info()
```

The ouliers has been successfully recoded with the median value of each column.
