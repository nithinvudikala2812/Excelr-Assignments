#NLP AND NAIVE BAYES

```python
import nltk
nltk.download('all')
```

#Data Exploration and Preprocessing

```python
import pandas as pd
import numpy as np
df=pd.read_csv("blogs.csv")
df
```

```python
#tokenization
from nltk.tokenize import word_tokenize

df['Tokenized_Data'] = df['Data'].apply(word_tokenize)
display(df.head())
```

```python
#normalization:convert all words to lowercase
df['Normalized_Data'] = df['Tokenized_Data'].apply(lambda x: [word.lower() for word in x])
display(df.head())
```

```python
#removing punctuation
import string
df['Punctuation_Removed'] = df['Normalized_Data'].apply(lambda x: [word for word in x if word not in string.punctuation])
display(df.head())
```

```python
#removing stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
df['Stop_Words_Removed'] = df['Punctuation_Removed'].apply(lambda x: [word for word in x if word not in stop_words])
display(df.head())
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
df['Lemmatized_Data'] = df['Stop_Words_Removed'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
display(df.head())
```

```python
#feature extraction
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
corpus = [' '.join(words) for words in df['Lemmatized_Data']]
X = vectorizer.fit_transform(corpus)
feature_names = vectorizer.get_feature_names_out()
df_features = pd.DataFrame(X.toarray(), columns=feature_names)
display(df_features.head())
```

#Naive Bayes Model for Text Classification

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_features, df['Labels'], test_size=0.2, random_state=42)
print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

accuracy_score=accuracy_score(y_test,y_pred)
print("Accuracy Score:",accuracy_score)
```

#Sentiment Analysis

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
df['Sentiment_Score'] = df['Data'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
df['Sentiment_Label'] = df['Sentiment_Score'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))
display(df.head())
```
